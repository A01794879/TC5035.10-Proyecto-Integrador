# Antecedentes

**MiCoto.mx** es una empresa tecnológica en modalidad *SaaS* (*Software as a Service*), especializada en la **administración digital de condominios residenciales**. Fundada para transformar la gestión tradicional de administradores y residentes, su plataforma opera en ciudades como **Guadalajara**, **Quintana Roo** y **Estado de México**, ofreciendo:  
- Registro de pagos y emisión de recibos.  
- Notificaciones automatizadas.  
- Coordinación de mantenimientos.  
- Centralización de datos financieros y operativos.  

### Retos Actuales  
A pesar de su base sólida, la empresa enfrenta tres desafíos críticos en **datos e inteligencia artificial**:  

1. **Gestión de morosidad**:  
   - Falta de modelos predictivos para anticipar impagos en cuotas de mantenimiento.  
   - Los administradores actúan de forma reactiva, generando problemas de liquidez y conflictos.  

2. **Acceso a información**:  
   - Consultas manuales en reportes históricos, limitando la experiencia de usuarios no técnicos.  
   - Toma de decisiones ralentizada por la navegación no intuitiva.  

3. **Interpretación de reportes**:  
   - Análisis financiero y de desempeño requiere expertise, excluyendo a usuarios que necesitan insights automáticos.  

### Objetivo del Proyecto  
Transformar la plataforma hacia un sistema **inteligente y analítico** mediante:  
- 🔮 **Modelos supervisados**: Predicción de morosidad.  
- 💬 **Procesamiento de Lenguaje Natural (PLN)**: Consultas conversacionales a la base de datos.  
- 🤖 **Chatbots**: Mejora en la interacción con usuarios.  
- 📊 **Generación automática de insights**: Visualizaciones y alertas para decisiones rápidas.  

### Datos Disponibles  
El análisis exploratorio utiliza un dataset estructurado con:  
- Historial de pagos por unidad habitacional.  
- Fechas de corte, montos y frecuencia de transacciones.  
- Políticas de cobro por condominio.  
- Interacciones en plataforma (para identificar patrones de uso).  

## Analisis de variables

````python
# Configuración inicial
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)
pd.set_option('display.max_columns', None)

# Cargar los datos
df = pd.read_excel(r"C:\Users\AMUNGUIA\Desktop\base propuesta micoto.xlsx")

# Función para convertir fechas mixtas
def convert_mixed_dates(col):
    if col.dtype == object:
        # Convertir a datetime donde sea posible, otros a NaN
        return pd.to_datetime(col, errors='coerce')
    return col

# Convertir columnas de fecha
date_cols = ['fecha_creacion', 'fecha_pago', 'hasta el mes']
for col in date_cols:
    if col in df.columns:
        df[col] = convert_mixed_dates(df[col])

# Función mejorada para analizar cada variable
def analyze_variable(df, column_name):
    print(f"\n{'='*50}")
    print(f"Análisis de la variable: {column_name}")
    print(f"{'='*50}")
    
    # Información básica
    print("\nInformación básica:")
    print(f"Tipo de dato: {df[column_name].dtype}")
    print(f"Valores únicos: {df[column_name].nunique()}")
    print(f"Valores nulos: {df[column_name].isnull().sum()} ({df[column_name].isnull().mean()*100:.2f}%)")
    
    # Manejar diferentes tipos de datos
    if np.issubdtype(df[column_name].dtype, np.number):
        # Análisis para variables numéricas
        print("\nEstadísticas descriptivas:")
        print(df[column_name].describe())
        
        # Gráfico de distribución
        plt.figure(figsize=(12, 6))
        sns.histplot(df[column_name], kde=True, bins=30)
        plt.title(f'Distribución de {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Frecuencia')
        plt.show()
        
        # Boxplot
        plt.figure(figsize=(8, 6))
        sns.boxplot(y=df[column_name])
        plt.title(f'Boxplot de {column_name}')
        plt.show()
    
    elif pd.api.types.is_datetime64_any_dtype(df[column_name]):
        # Análisis para variables de fecha
        print("\nRango de fechas:")
        print(f"De {df[column_name].min()} a {df[column_name].max()}")
        
        # Evolución temporal
        plt.figure(figsize=(12, 6))
        df[column_name].dt.to_period('M').value_counts().sort_index().plot()
        plt.title(f'Evolución temporal de {column_name}')
        plt.xlabel('Fecha')
        plt.ylabel('Frecuencia')
        plt.show()
    
    else:
        # Análisis para variables categóricas
        print("\nValores más frecuentes:")
        print(df[column_name].value_counts().head(10))
        
        # Convertir a string para evitar problemas con tipos mixtos
        temp_series = df[column_name].astype(str)
        
        # Gráfico de barras para las categorías principales
        plt.figure(figsize=(12, 6))
        top_categories = temp_series.value_counts().head(10)
        sns.barplot(x=top_categories.index, y=top_categories.values)
        plt.title(f'Top 10 categorías de {column_name}')
        plt.xticks(rotation=45)
        plt.xlabel(column_name)
        plt.ylabel('Frecuencia')
        plt.show()

# Análisis de cada variable
for column in df.columns:
    try:
        analyze_variable(df, column)
    except Exception as e:
        print(f"\nError al analizar {column}: {str(e)}")
        continue

# Análisis específico para fechas de pago
if 'fecha_pago' in df.columns and 'fecha_creacion' in df.columns:
    # Calcular días hasta el pago solo para registros con ambas fechas válidas
    valid_dates = df[['fecha_pago', 'fecha_creacion']].notna().all(axis=1)
    df.loc[valid_dates, 'dias_pago'] = (df.loc[valid_dates, 'fecha_pago'] - df.loc[valid_dates, 'fecha_creacion']).dt.days
    
    print("\nAnálisis de días hasta el pago:")
    print(df['dias_pago'].describe())
    
    plt.figure(figsize=(12, 6))
    sns.histplot(df['dias_pago'].dropna(), bins=30, kde=True)
    plt.title('Distribución de días hasta el pago')
    plt.xlabel('Días hasta el pago')
    plt.ylabel('Frecuencia')
    plt.show()
    
    # Relación entre monto y días de pago
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='dias_pago', y='monto', data=df)
    plt.title('Relación entre monto y días hasta el pago')
    plt.xlabel('Días hasta el pago')
    plt.ylabel('Monto')
    plt.show()

# Análisis de morosidad
if 'moros' in df.columns:
    print("\nAnálisis de morosidad:")
    print(df['moros'].value_counts())
    
    plt.figure(figsize=(12, 6))
    sns.countplot(x='moros', data=df)
    plt.title('Distribución de morosidad')
    plt.xlabel('Número de moras')
    plt.ylabel('Frecuencia')
    plt.show()
    
    # Relación entre morosidad y monto
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='moros', y='monto', data=df)
    plt.title('Relación entre morosidad y monto')
    plt.xlabel('Número de moras')
    plt.ylabel('Monto')
    plt.show()

# Análisis por tipo de pago
if 'tipo' in df.columns:
    print("\nAnálisis por tipo de pago:")
    print(df['tipo'].value_counts())
    
    plt.figure(figsize=(12, 6))
    sns.countplot(x='tipo', data=df)
    plt.title('Distribución por tipo de pago')
    plt.xlabel('Tipo de pago')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.show()
    
    # Monto promedio por tipo
    plt.figure(figsize=(12, 6))
    sns.barplot(x='tipo', y='monto', data=df, estimator=np.mean)
    plt.title('Monto promedio por tipo de pago')
    plt.xlabel('Tipo de pago')
    plt.ylabel('Monto promedio')
    plt.xticks(rotation=45)
    plt.show()

# Análisis por estatus de pago
if 'Estatus pago' in df.columns:
    print("\nAnálisis por estatus de pago:")
    print(df['Estatus pago'].value_counts())
    
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Estatus pago', data=df)
    plt.title('Distribución por estatus de pago')
    plt.xlabel('Estatus de pago')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.show()

# Evolución temporal de montos
if 'fecha_creacion' in df.columns and 'monto' in df.columns:
    df['mes_creacion'] = df['fecha_creacion'].dt.to_period('M')
    
    plt.figure(figsize=(12, 6))
    df.groupby('mes_creacion')['monto'].mean().plot()
    plt.title('Evolución del monto promedio por mes')
    plt.xlabel('Mes')
    plt.ylabel('Monto promedio')
    plt.show()

# Análisis por casa
if 'casa' in df.columns:
    print("\nTop 10 casas con más pagos:")
    print(df['casa'].value_counts().head(10))
    
    plt.figure(figsize=(12, 6))
    df['casa'].value_counts().head(10).plot(kind='bar')
    plt.title('Top 10 casas con más pagos')
    plt.xlabel('Casa')
    plt.ylabel('Número de pagos')
    plt.xticks(rotation=45)
    plt.show()

# Análisis por coto
if 'coto' in df.columns:
    print("\nDistribución por coto:")
    print(df['coto'].value_counts())
    
    plt.figure(figsize=(12, 6))
    sns.countplot(x='coto', data=df)
    plt.title('Distribución por coto')
    plt.xlabel('Coto')
    plt.ylabel('Frecuencia')
    plt.show()
    
    # Monto promedio por coto
    plt.figure(figsize=(12, 6))
    sns.barplot(x='coto', y='monto', data=df, estimator=np.mean)
    plt.title('Monto promedio por coto')
    plt.xlabel('Coto')
    plt.ylabel('Monto promedio')
    plt.show()
````

# Conclusiones

- **Datos faltantes**:  
  - No hay valores faltantes críticos, salvo en `fecha_pago`, lo que podría indicar transacciones pendientes o registros no actualizados.  
  - Valores categóricos con etiquetas ambiguas requieren estandarización.  

- **Distribuciones y valores atípicos**:  
  - `monto` y `moras` presentan distribuciones sesgadas (derecha) con valores extremos.  
  - `monto` tiene transacciones ocasionales muy superiores al promedio, lo que puede afectar modelos sensibles a escala.  
  - `moras` y `numero_moras_maxima` muestran alta concentración en valores bajos (0 o 1), pero con un grupo recurrente de morosidad.  

- **Patrones predictivos**:  
  - `tipo`:  
    - Categorías dominantes: "afavor" (casi sin morosidad) y "mantenimiento" (asociado a morosidad).  
  - `status`:  
    - Estados "pendiente" y "atrasado" son predictores clave de morosidad.  
    - "pagado" o "condonado" no presentan morosidad por definición.  
  - Correlación positiva entre `numero_moras_maxima` y `moras`, confirmando su relación con el comportamiento de pago.  

- **Desbalance de clases**:  
  - Variable `Moroso` tiene desequilibrio moderado (mayoría "NO").  
  - Estrategias recomendadas: SMOTE, ponderación de clases o muestreo estratificado.  

- **Manejo de variables**:  
  - **Categóricas**:  
    - `tipo`, `status`, `Estatus_pago` y `Moroso` son ideales para *one-hot encoding*.  
  - **Cardinalidad alta**:  
    - `coto` y `casa` tienen alta cardinalidad. Excluir `casa` o codificar `coto` según morosidad promedio.  
  - **Fechas**:  
    - `fecha_creacion` y `fecha_pago` permiten analizar estacionalidad (ej. morosidad en meses específicos).
